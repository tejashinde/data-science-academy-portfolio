```python
import numpy as np # linear algebra
import pandas as pd # data processing
import warnings
warnings.filterwarnings('ignore')
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('output-onlinecsvtools.csv')
```

```python
df.shape
```

```python
df = df.drop('2', axis = 1) 
```

```python
df.columns = ['p1','p2']
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler, normalize
scaler = StandardScaler() 
scaled_df = scaler.fit_transform(df) 
  
# Normalizing the Data 
normalized_df = normalize(scaled_df) 
  
# Converting the numpy array into a pandas DataFrame 
normalized_df = pd.DataFrame(normalized_df)

```

```python
normalized_df.columns = ['P1', 'P2']
```

```python
from sklearn.cluster import SpectralClustering
```

```python
spectral_model = SpectralClustering(n_clusters = 4, affinity ='nearest_neighbors') 
  

spectral_model.fit_predict(normalized_df)
```

```python

# Visualizing the clustering 
plt.scatter(normalized_df['P1'], normalized_df['P2'],  
           c = SpectralClustering(n_clusters = 4, affinity ='nearest_neighbors') .fit_predict(normalized_df), cmap =plt.cm.RdYlGn) 
plt.show()
```
